```python
import pandas as pd

move_df = pd.read_csv('/kaggle/input/predicting-pro-chess-moves/output.csv')
```

```python
move_df_values = move_df.values
processed_fens = []
moves = []
for i in range(0, len(move_df)):
    if(move_df_values[i][0].split(" ")[1] == 'w'):
        processed_fens.append(move_df_values[i][0].split(" ")[0])
        moves.append(move_df_values[i][1])
```

```python
white_move_df = pd.DataFrame()
white_move_df['FEN'] = processed_fens
white_move_df['Move'] = moves
white_move_df
```

```python
pre_df = pd.read_feather('/kaggle/input/csv-from-json/NEW_all_unprocessed_chess_dataset.feather')
pre_df
```

```python
index_numbers = pre_df[pre_df['current_move'] == 'b'].index 
pre_df = pre_df.drop(index_numbers) 
pre_df
```

```python
binary_encoder = [
"100000000000", # r
"010000000000", # n 
"001000000000", # b
"000100000000", # q
"000010000000", # k
"000001000000", # p
"000000100000", # P
"000000010000", # R
"000000001000", # N
"000000000100", # B
"000000000010", # Q
"000000000001", # K
"000000000000"
]

pieces_encoder = ['r', 'n', 'b', 'q', 'k', 'p', 'P', 'R', 'N', 'B', 'Q', 'K']
piece_values   = [-5,   -3,  -3,  -9,   0,  -1,   1,   5,   3,   3,   9,   0]
```

```python
pre_df = pre_df.reset_index()
del pre_df['index']
pre_df
```

```python
# 1 - pre_df[pre_df['pv_type'] == 'mate']

scaled_evaluation_values = []

pre_df_values = pre_df.values
for i in range(0, len(pre_df_values)):
    if(pre_df_values[i][2] == 'mate'):
        if(float(pre_df_values[i][3]) > 0): # value of 0.05 means 12 moves away from checkmate
#             print(((1 - (float(pre_df_values[i][3])))  * 10) + 20) # 0.12 -> 28.80
            scaled_evaluation_values.append(((1 - (float(pre_df_values[i][3])))  * 10) + 20)
        else:
#             print(-1 * ((1 - abs(float(pre_df_values[i][3])))  * 10) - 20) # -0.06 -> -29.4
            scaled_evaluation_values.append(-1 * ((1 - abs(float(pre_df_values[i][3])))  * 10) - 20)
    else:
        scaled_evaluation_values.append(pre_df_values[i][3])
```

```python
pre_df['evaluation_value'] = scaled_evaluation_values
del pre_df['pv_type']
del pre_df['current_move']
pre_df
```

```python
pre_df.to_feather("/kaggle/working/white_move_only1.feather")  
```
