```python
length = len(df)

train_length = round(length * .7)
val_length = round(length * .9)

train_df = df[:train_length]
val_df = df[train_length:val_length]
test_df = df[val_length:]

X_train = train_df.drop('evaluation_value',axis=1)
y_train = train_df['evaluation_value']

X_val = val_df.drop('evaluation_value',axis=1)
y_val = val_df['evaluation_value']

X_test = test_df.drop('evaluation_value',axis=1)
y_test = test_df['evaluation_value']

input_shape = [X_train.shape[1]]
input_shape
```

```python
X_train = np.array(X_train).reshape(len(X_train), 8, 8)
X_val = np.array(X_val).reshape(len(X_val), 8, 8)
X_test = np.array(X_test).reshape(len(X_test), 8, 8)
```

```python
from keras.models import Sequential
from keras.layers import Dense, Conv1D, Flatten
#create model
model = Sequential()
#add model layers
model.add(Conv1D(64, kernel_size=3, activation='relu', input_shape=(8, 8)))
model.add(Conv1D(32, kernel_size=3, activation='relu'))
model.add(Flatten())
model.add(Dense(1, activation='softmax'))
```

```python
model.compile(optimizer='adam', loss='mse', metrics=['mae'])
```

```python
model.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=1)
model.save('/kaggle/working/cnntest.keras')
```
